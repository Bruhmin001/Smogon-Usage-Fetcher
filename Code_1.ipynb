```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('master_sheet.csv')
```

```python
pkm = input("Enter the Name of the pokemon to get the info")
pkm = str.upper(pkm)
data = []
print(pkm)
for index, row in df.iterrows():
    if row['Name'] == pkm:
        stat = row.values.tolist()
        data.append(stat)
print(data)
    

```

```python
rank = [] 
name =[]
usage_perc =[]
raw =[]
raw_perc =[]
real =[]
real_perc=[]
period = []
for i in range (0,len(data)):
        rank.append(data[i][2])
        name.append(data[i][3])
        usage_perc.append(data[i][4])
        raw.append(data[i][5])
        raw_perc.append(data[i][6])
        real.append(data[i][7])
        real_perc.append(data[i][8])
        period.append(data[i][1])
```

```python

df = pd.DataFrame({'Month':period,'Rank':rank,'Name':name,'Usage %':usage_perc,'Raw usage':raw,'Raw %':raw_perc,'Real usage':real,'Real %':real_perc})
df.to_csv("Tableau.csv")
df
```
